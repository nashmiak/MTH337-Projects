## Part 1

This part includes the depiction of the frequency of the 20 most common words that appear in each of the four texts, as shown on a bar graph. The words, along with their frequencies, are shown in a list comprised from the definition functions `finaltext(x)`, `counttext(x)`, `sortlist(k)`, and `sort(countlist)`.

```python
#Importing the necessary commands
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Extraction of each of the necessary text files to be used throughout the report
text1=requests.get('http://www.acsu.buffalo.edu/~scassani/MTH337/_downloads/58ca473582a8ec156f1a18df7f24ee39/around_world.txt').text
text2=requests.get('http://www.acsu.buffalo.edu/~scassani/MTH337/_downloads/31144e46885f987b907024c7c9ce24c9/republic.txt').text
text3=requests.get('http://www.acsu.buffalo.edu/~scassani/MTH337/_downloads/09e65fe685278fdab4a0c0d29fa09653/walden.txt').text
text4=requests.get('http://www.acsu.buffalo.edu/~scassani/MTH337/_downloads/ad16bce3cc209d74e598118909a2f78d/three_musketeers.txt').text
```

These defintions correlate with each of the texts that will be analyzed in the report. `text1` correlates with *Around the World in Eighty Days*, `text2` correlates with *The Republic*, `text3` correlates with *Walden*, and `text4` correlates with *The Three Musketeers*.

```python
def finaltext(x):
    x=x.lower() #makes the text in the file all lowercase to easily identify elements in the entire text
    skip='":;<>,.()[]-_!?\r\n' #items that will be skipped in the text
    for k in skip:
        x=x.replace(k,' ')
    y=x.split(' ') 
    vlist=[k for k in y if k!=''] #list comprehension to return the cleaned text within a list
    
    return vlist
```

This definition function,`finaltext(x)`, replaces the unnecessary symbols in the extracted text with an empty space, and returns each word of the text separated in a list.

```python
def counttext(x):
    countdict={}
    for i in x:
        if i in countdict:
            countdict[i]+=1
        else: 
            countdict[i]=1
#includes the count of the words from the text in a dictionary     
    return countdict
```

This definition function,`counttext(x)`, creates a dictionary that holds the count of the characters found within a text file.

```python
def sortlist(k):
    return k[1] #returns the first element in the list
```

This definition function,`sortlist(k)`, returns the first element within a list of sorted items, and will be used for the `sort(countlist)` function as the key.

```python
def sort(countlist):
    return sorted(countlist,key=sortlist,reverse=True)
```

This definition function, `sort(countlist)`, returns the words in the `sortlist(k)` function in decreasing numerical order based on how frequently they appear in the text.

```python
#Applying the formulas from above to the first text in order to create a list and bar graph of the most frequent words.
textk1=finaltext(text1)
textk1count=counttext(textk1)
countlist=list(textk1count.items())
clean1=sort(countlist)
print(clean1[:20]) #Creates a list of the 20 words most frequently appearing in the first text.

word=[k[0] for k in clean1[:20]] #Takes into consideration the word
frequency=[k[1] for k in clean1[:20]] #Takes into consideration the frequency of the word appearing in the text
words1=plt.bar(range(1,21),frequency,color='r') #Creation of the bar graph for text1
plt.title('Around the World in Eighty Days Frequent Words',fontsize=12)
plt.xlabel('Word',fontsize=10)
plt.ylabel('Frequency',fontsize=10)
plt.show()
```

This depicts the 20 most common words from *Around the World in Eighty Days* in a list and how frequently they appear in the text as shown on a bar graph.

```python
#Applying the formulas from above to the second text in order to create a list and bar graph of the most frequent words.
textk2=finaltext(text2)
textk2count=counttext(textk2)
countlist=list(textk2count.items())
clean2=sort(countlist) #Creates a list of the 20 words most frequently appearing in the second text.
print(clean2[:20]) 

word=[k[0] for k in clean2[:20]]
frequency=[k[1] for k in clean2[:20]]
words2=plt.bar(range(1,21),frequency,color='y')  #Creation of the bar graph for text2
plt.title('The Republic Frequent Words',fontsize=12)
plt.xlabel('Word',fontsize=10)
plt.ylabel('Frequency',fontsize=10)
plt.show()
```

This depicts the 20 most common words from *The Republic* in a list and how frequently they appear in the text as shown on a bar graph.

This part includes the analyzation of the relationship between the words in terms of the function `log(x)` and the actual frequencies of the words in terms of the function `log(y)`. By taking into consideration various correlations between x and y, such as x vs. log(y), or log(x) vs. y, and finally, log(x) vs. log(y), various conclusions can be drawn. The most significant relationship that will be focused on for this project is the association between log(x) vs. log(y) corresponding to the most commons words throughout the entirety of the text, along with their frequencies.

```python
#Defines the x and y values that will be used to plot log(x) vs. log(y) for the first text
x1=np.arange(1,len(clean1)+1)
y1=np.array([k[1] for k in clean1])
```

```python
#Defines the x and y values that will be used to plot log(x) vs. log(y) for the second text
x2=np.arange(1,len(clean2)+1)
y2=np.array([k[1] for k in clean2])
```

```python
#Defines the x and y values that will be used to plot log(x) vs. log(y) for the third text
x3=np.arange(1,len(clean3)+1)
y3=np.array([k[1] for k in clean3])
```

```python
#Defines the x and y values that will be used to plot log(x) vs. log(y) for the fourth text
x4=np.arange(1,len(clean4)+1)
y4=np.array([k[1] for k in clean4])
```

These values of x and y in relation to the associated texts will be used to create a graph that plots the logarithmic frequencies of the values x and y.

```python
#Plots the log(x) vs. log(y) functions for the four texts on the same graph to clearly see correlation
plt.figure(figsize=(5,5))
plt.plot(np.log(x1),np.log(y1),'r')
plt.plot(np.log(x2),np.log(y2),'y')
plt.plot(np.log(x3),np.log(y3),'g')
plt.plot(np.log(x4),np.log(y4),'m')
plt.xlabel('Log(x_i)',fontsize=15)
plt.ylabel('Log(y_i)',fontsize=15)
plt.title('Log(x_i) vs. Log(y_i) Frequencies',fontsize=18)
plt.show()
```

The relationship that represents the decay of the most frequent word as shown within each text is very similar. The logarithmic function is used to clearly depict the relationship between the word frequencies within each of these specific texts all published in different years by different people at various places in the world. This can be shown when looking at the slope, $\left(\frac{1}{x^n}\right)$ , where n represents the value of the slope. As can be seen, the publisher, place, or date of publication of each of the texts seem to not have any effect on the word usage of these different time periods or people, because it all translates into words commonly used in the english language today.

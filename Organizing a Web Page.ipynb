## Project 8

This project included the retrieval of contact information from each of the faculty listed within the University at Buffalo's School of Pharmacy homepage. The goal of this project is to apply the different characters used to identify each section within the given webpage including the first and last name, address of the faculty member's office, their work phone number, and their University at Buffalo email address, while taking into consideration the different versions each of these sections are represented.

```python
import requests
import re
```

```python
pharma=requests.get('http://pharmacy.buffalo.edu/faculty-staff.html').text
links=re.findall(r'"facultyindexname" href="(.*?)">',pharma)
```

These defintions, `pharma` and `links`, include using the import functions to retrieve and organize the data by separating each of the faculty's webpages to be worked off of.

```python
#Application of the separate functions that will be used to acquire the necessary information from the 
# contact page of all of the faculty within UB's School of Pharmacy. 
for n in links:
    site=requests.get(n).text
#Request function that will retrieve each individual faculty member's data from the given site.    
        
    name=re.findall(r'<span id="lblFacultyName">(.*?)</span>',site)[0]
    print(name)
#This function returns the full name of the faculty members on the webpage.    
        
    info=re.findall(r'"lblAcademicOfficeAddressMobile">(.*?)</a><br/>',site)
    address=re.findall(r'(.*?)Phone:',info[0])
    if len(address)==0:
        address='No Address'
    else:
        address=address[0].replace('<br/>',' ')
    print(address)
#This function returns the address of the faculty member's office, while taking into consideration the lack of an office
# address to be easily identified as `No Address`.
        
    phone=re.findall(r'Phone: (.*?) <br/>',info[0])
    if len(phone)==0:
        phone='No Phone'
    else:
        phone=phone[0] #Only takes into consideration the first phone number seen prior to the faculty member's name.
        k=re.findall(r'\d',phone)
        phone='('+''.join(k[:3])+') '+ (''.join(k[3:6])) + '-'+(''.join(k[6:10])) #Formats the phone number on the webpage into the standard form.
    print(phone)
#This function returns the work phone numbers for each of the faculty members on the webpage, while organizing the numbers
# to a standard format, and identifying the lack of a phone number as `No Phone`.
        
    email=re.findall(r'href="mailto:(.*?)">',info[0])
    if len(email)==0:
        email='No Email'
    else:
        email=email[0]
    print(email)
    print(' ')
#This function returns the work emails of each of the faculty members on the webpage, while taking into consideration
# the lack of an email to be easily identified as 'No Email'.
```

This `for` loop includes various commands that return the given output, which includes the retrieval of each member on the faculty website for UB's School of Pharmacy showing their contact information in an easily readable format alphabetically. The code is able to handle issues such as missing information, or different formatting styles of certain information as shown on the website to make the directory all in one spot rather than individual links.
